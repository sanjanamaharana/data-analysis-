```python
# Instance Method
# An instance method works with object (instance) data.

# Example 1

class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

s = Student("Sanjana")
s.display()
```

```python
# Example 2
class Car:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print("Brand:", self.brand)

c = Car("Toyota")
c.show_brand()
```

```python
# Example 3

class Employee:
    def __init__(self, salary):
        self.salary = salary

    def show_salary(self):
        print("Salary:", self.salary)

e = Employee(30000)
e.show_salary()
```

```python
# Example 4

class Book:
    def __init__(self, title):
        self.title = title

    def display_title(self):
        print("Book:", self.title)

b = Book("Python Basics")
b.display_title()
```

```python
# Example 5

class Mobile:
    def __init__(self, model):
        self.model = model

    def show_model(self):
        print("Model:", self.model)

m = Mobile("Samsung")
m.show_model()
```

```python
# Example 6

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("Area:", 3.14 * self.radius * self.radius)

c = Circle(5)
c.area()
```

```python
# 2. Class Method (Uses @classmethod and cls)

# A class method works with class variables.

# Example 1
class Student:
    school = "ABC School"

    @classmethod
    def show_school(cls):
        print(cls.school)

Student.show_school()
```

```python
# Example 2
class Employee:
    company = "Infosys"

    @classmethod
    def company_name(cls):
        print(cls.company)

Employee.company_name()
```

```python
# Example 3
class Car:
    wheels = 4

    @classmethod
    def show_wheels(cls):
        print("Wheels:", cls.wheels)

Car.show_wheels()
```

```python
# Example 4
class College:
    name = "XYZ College"

    @classmethod
    def display(cls):
        print(cls.name)

College.display()
```

```python
# Example 5
class Bank:
    bank_name = "SBI"

    @classmethod
    def show_bank(cls):
        print(cls.bank_name)

Bank.show_bank()
```

```python
# Example 6
class Laptop:
    brand = "Dell"

    @classmethod
    def show_brand(cls):
        print(cls.brand)

Laptop.show_brand()
```

```python
# 3. Static Method (Uses @staticmethod)

# A static method does not use self or cls. It performs a general utility task.

# Example 1
class Math:
    @staticmethod
    def add(a, b):
        print(a + b)

Math.add(10, 20)
```

```python
# Example 2
class Calculator:
    @staticmethod
    def square(n):
        print(n * n)

Calculator.square(6)
```

```python
# Example 3
class Temperature:
    @staticmethod
    def celsius_to_fahrenheit(c):
        print((c * 9/5) + 32)

Temperature.celsius_to_fahrenheit(30)
```

```python
# Example 4
class Message:
    @staticmethod
    def greet():
        print("Welcome!")

Message.greet()
```

```python
# Example 5
class Number:
    @staticmethod
    def even_or_odd(n):
        if n % 2 == 0:
            print("Even")
        else:
            print("Odd")

Number.even_or_odd(9)
```

```python
# Example 6
class Circle:
    @staticmethod
    def pi():
        print(3.14159)

Circle.pi()
```

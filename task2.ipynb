```python
# Q30.
# Create a dictionary where:
# - Key = Student Name
# - Value = Marks
# Display all students who scored more than 80 marks.

students = {
    "Sanjana": 85,
    "Priya": 90,
    "Riya": 78,
    "Anjali": 88,
    "Pooja": 75
}

print("Students who scored more than 80 marks:")

for name, marks in students.items():
    if marks > 80:
        print(name, "-", marks)
```
